```python
import pandas as pd
import matplotlib.pyplot as plt

stats = ['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']

for stat in stats:
    # Create a figure with two subplots (side-by-side)
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # 1 row, 2 columns

    # Original histogram (left subplot)
    df[stat].plot(kind='hist', colormap='viridis', title=f"{stat} values of Pokemon", ax=axes[0])
    axes[0].set_xlabel(f"{stat.capitalize()} Values")

    # Calculate IQR and bounds
    Q1 = df[stat].quantile(0.25)
    Q3 = df[stat].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filtered histogram (right subplot)
    filtered_df = df.loc[(df[stat] < upper_bound) & (df[stat] > lower_bound)]
    filtered_df[stat].plot(kind='hist', colormap='viridis', title=f"{stat} values of Pokemon (outliers removed)", ax=axes[1])
    axes[1].set_xlabel(f"{stat} values")

    # Find and display outliers
    outlier_df = df.loc[(df[stat] > upper_bound) | (df[stat] < lower_bound), ['name', stat]].sort_values(by=[stat], ascending=False)
    if not outlier_df.empty:
        print(f"Outliers for {stat}:")
        print(outlier_df)
    else:
        print(f"No outliers found for {stat}")

    plt.tight_layout()  # Adjust spacing between subplots
    plt.show()
```

```python
# Assuming 'df' is your DataFrame
number_df = df.select_dtypes(include=np.number)  # Corrected: include=np.number
correlation_df = number_df.iloc[:, 0:20].corr()

# Create a mask for correlations below 0.7
mask = np.abs(correlation_df) < 0.6

# Create the heatmap with the mask and adjust color for visibility
plt.figure(figsize=(12, 10))  # Adjust figure size if needed
sns.heatmap(
    correlation_df,
    cmap="coolwarm",
    annot=True,
    mask=mask,
    vmin=-1,
    vmax=1,
    linewidths=0.5,
    linecolor="black",
)  # Annotate cells, set min/max values, add lines

plt.title("Correlation Heatmap (Correlation >= 0.7)")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assuming your DataFrame is already loaded as 'df'
# ... (Your code to load the DataFrame) ...

# --- Outlier Removal (for both weight_kg and height_m) ---

# Weight Outlier Removal
Q1_weight = df['weight_kg'].quantile(0.25)
Q3_weight = df['weight_kg'].quantile(0.75)
IQR_weight = Q3_weight - Q1_weight
lower_bound_weight = Q1_weight - 1.5 * IQR_weight
upper_bound_weight = Q3_weight + 1.5 * IQR_weight

# Height Outlier Removal
Q1_height = df['height_m'].quantile(0.25)
Q3_height = df['height_m'].quantile(0.75)
IQR_height = Q3_height - Q1_height
lower_bound_height = Q1_height - 1.5 * IQR_height
upper_bound_height = Q3_height + 1.5 * IQR_height

# Filter the DataFrame, keeping only data within the bounds for BOTH weight and height
df_no_outliers = df.loc[
    (df['weight_kg'] < upper_bound_weight) & (df['weight_kg'] > lower_bound_weight) &
    (df['height_m'] < upper_bound_height) & (df['height_m'] > lower_bound_height)
]

# --- 3D Plot Creation ---

fig = plt.figure(figsize=(10, 8))  # Adjust figure size as needed
ax = fig.add_subplot(111,projection='3d')

# Scatter plot in 3D
ax.scatter(df_no_outliers['weight_kg'], df_no_outliers['height_m'], df_no_outliers['total_points'],
           c=df_no_outliers['total_points'], cmap='viridis', alpha=0.6)  # Color by 'total_points'

# Set labels and title
ax.set_xlabel('Weight (kg)')
ax.set_ylabel('Height (m)')
ax.set_zlabel('Total Points')
ax.set_title('3D Scatter Plot of Weight, Height, and Total Points (Outliers Removed)')

# Add a colorbar
cbar = fig.colorbar(ax.collections[0], ax=ax, shrink=0.5, aspect=10)
cbar.set_label('Total Points')

plt.show()
```
